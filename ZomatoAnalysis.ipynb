```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Zomato data .csv',encoding = 'unicode_escape')
df
```

```python
df
```

```python
# data cleaning = to clean rate column '/'
```

```python
def handlerate(value):
    value=str(value).split("/")
    value=value[0];
    return float(value)
df['rate']=df['rate'].apply(handlerate)
df
```

```python
df.info()
```

```python
# q.1)what type of restaurant do the majority of customers orders from?
```

```python
df.head()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('type of restaurent')
```

```python
# conclusion - majority of the restaurant falls in dinning category
```

```python
# q.2) How many votes has each type of restaurant received from customers?
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,c='green',marker='o')
plt.xlabel('type of restaurant',c='red',size=20)
plt.ylabel('votes',c='red',size=20)
```

```python
df.head()
```

```python
# conclusion - dinning restaurant has received maximum votes
```

```python
# q.3) What are the ratings that the majority of restaurants have received?
```

```python
plt.hist(df['rate'],bins = 5)
plt.title('rating distribution')
plt.show()
```

```python
# conclusion - majority of restaurant received ratings from 3.5 to 4
```

```python
# q.5) Which mode (online or offline) has received the maximum rating?
```

```python
df
```

```python
sns.boxplot(data=df,x='online_order',y='rate')
plt.figure(figsize=(6,6))
```

```python
# conclusion = online orders receives high rating than offline orders
```

```python
# q.6) Which type of restaurant received more offline orders?
```

```python
# gere we use pivote_table function
pivot_table = df.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu',fmt='d')
plt.title('Heatmap')
plt.xlabel('online_order')
plt.ylabel('listen_in(type)')
plt.show()
```

```python
# conclusion = Dinning restaurant primarily accept offline orders,whereas cafes primarily receive online orders. This suggests
# that clients prefer orders in person in restaurant but prefer online ordering at cafes.
```
